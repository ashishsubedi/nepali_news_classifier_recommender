### Objective
To convert the raw data to csv format for easy use

```python
import os
from tqdm import tqdm
```

```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import nltk
```

```python
DATA_DIR = os.path.abspath('./datasets/20_news')
OUTPUT_DIR = os.path.join('./datasets/output')
```

### Converting raw data to csv format
Data Format: id, class, text

```python
# RAW_DATA_DIR = os.path.join(DATA_DIR,'raw')
RAW_DATA_DIR = DATA_DIR
OUTPUT_FILENAME = 'data_2.csv'
```

```python
classes = os.listdir(RAW_DATA_DIR)
columns=['id','class','text']
```

```python
data_dict = {
    'id':[],
    'class':[],
    'text': []
}
```

```python

for label in classes:
    files = os.path.join(RAW_DATA_DIR,label)
    for f in tqdm(os.listdir(files)):
        id_ = f.split('.')[0]
        try:
            with open(os.path.join(files,f),encoding='utf-8') as content:
                text = content.read().strip()
            #Adding corresponding data to the dictionary
            data_dict['id'].append(id_)
            data_dict['class'].append(label)
            data_dict['text'].append(text)
        except:
            print(label,id_,"Failed to load")


            
```

```python
df = pd.DataFrame(data_dict,columns=columns,dtype=object)
df.dropna(inplace=True)

```

```python
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
df.dropna(subset = ["text"], inplace=True)
```

```python
df.reset_index(inplace=True,drop=True)
df
```

```python
df['id'] = df.index
df.head()
```

```python
df.loc[2679]
```

```python
df = df.set_index('id')
df.head()
```

```python
df.to_csv(os.path.join(OUTPUT_DIR,OUTPUT_FILENAME))
```
